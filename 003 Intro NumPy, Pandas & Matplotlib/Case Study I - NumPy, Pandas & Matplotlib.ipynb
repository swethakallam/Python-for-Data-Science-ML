# 15. Analyse various school outcomes in Tennessee using pandas.

```python
import pandas as pd
import numpy as np
import matplotlib
from numpy import mean
from numpy import std
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot


# Imported CSV file and read it with pd.read_csv
tensse_school = pd.read_csv('middle_tn_schools.csv')
print(tensse_school)

#Described the CSV file with .describe()
desc = tensse_school.describe()
print(desc)

# Extracted column of School_Rating and Reduced_Lunched
scl_rting_n_rdcd_lnch_column = tensse_school.iloc[0:348, [1,3]]
print(scl_rting_n_rdcd_lnch_column)

seed(1)
scl_rting_n_rdcd_lnch_column = 20 * randn(1000) + 100
data2 = scl_rting_n_rdcd_lnch_column + (10 * randn(1000) + 50)

print('scl_rting_n_rdcd_lnch_column: mean=%.3f stdv=%.3f' % (mean(scl_rting_n_rdcd_lnch_column), std(scl_rting_n_rdcd_lnch_column)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))

pyplot.scatter(scl_rting_n_rdcd_lnch_column, data2)
pyplot
```
